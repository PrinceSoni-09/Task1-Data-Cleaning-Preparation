##Import the Libraries

```python
import numpy as np
import pandas as pd
```

##Load the Dataset

```python
data = pd.read_excel('Dataset for Data Analytics.xlsx')
data.head()
```

##Information About the Dataset

```python
data.info()
```

##Handling the Null Values

```python
''' Null values of coupon code defines that
    there is no coupon code applied on that order by the customer
    so we handle it by filling the null values with NA '''

data['CouponCode'] = data['CouponCode'].fillna('NA')
print(data.isnull().sum())
```

#Check For Duplicates

```python
''' No duplicates are there in dataset
    So no need to handle them'''
data.duplicated().sum()
```

##Check For White Spaces

```python
#Select all the columns with categorical values
string_cols = data.select_dtypes(include='object')

#Removing the whitespaces from starting, ending, and between words
data[string_cols.columns] = string_cols.apply(lambda col: col.str.replace(r'\s+', '', regex=True).str.strip())
```

##Check For Date Format

```python
#To Check all the dates are in proper format or not
date_mask = pd.to_datetime(data['Date'], format='%Y-%m-%d', errors = 'coerce').notna()
print(data[~date_mask])
#All the dates are in proper format
data.head()
```

##Check For Numeric Precision (2 Decimals)

```python
numeric_cols = ['UnitPrice', 'TotalPrice']

NP_mask = data[numeric_cols].apply(
    lambda col: ((col * 100).round(6) % 1 == 0)
)

print(data.loc[~NP_mask.all(axis=1), numeric_cols])
#No entry in both the columns having more than 2 precision
```

##Final Check:
##1. Error Rate on OrderID
##2. Error Rate on Date Formats

```python
#Calculating Error Rate for OrderID

duplicate_count = data['OrderID'].duplicated().sum()
total_rows = len(data)
error_rate_OID = (duplicate_count / total_rows) * 100

#Calculating Error Rate on Date Formats

invalid_dates = (~date_mask).sum()
error_rate_DF = (invalid_dates / total_rows) * 100
print(f'{error_rate_OID: .2f}% Error Rate on Unique Identifiers')
print(f'{error_rate_DF: .2f}% Error Rate on Date Formats')
```
